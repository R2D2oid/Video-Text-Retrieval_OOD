## Load MSRVTT dataset

```python
import torch
from torchvision import transforms
from msrvtt_dataset import MSRVTTDataset as MSRVTT
from msrvtt_dataset import Standardize_VideoSentencePair, ToTensor_VideoSentencePair
from utils.train_utils import get_dataloader

repo_dir = '/usr/local/extstore01/zahra/datasets/MSRVTT/'
video_feats_dir = f'{repo_dir}/feats/video/r2plus1d_TrainVal'
text_feats_path = f'{repo_dir}/feats/text/msrvtt_captions_np.pkl'
ids_path = f'{repo_dir}/TrainVal_videoid_sentid.txt'

dl_params = {'batch_size': 64,
             'shuffle': False,
             'num_workers': 1}
```

```python
dataset0 = MSRVTT(vid_feats_dir=video_feats_dir, txt_feats_path=text_feats_path, ids_path=ids_path, transform=None)
dataset_stats = dataset0.get_dataset_mean_std()
standardize = Standardize_VideoSentencePair(dataset_stats)
trnsfrm = transforms.Compose([standardize,
                              ToTensor_VideoSentencePair()])
dataset1 = MSRVTT(vid_feats_dir=video_feats_dir, txt_feats_path=text_feats_path, ids_path=ids_path, transform=trnsfrm)
```

```python
dataloader0 = torch.utils.data.DataLoader(dataset0, **dl_params)
dataloader1 = torch.utils.data.DataLoader(dataset1, **dl_params)
```
